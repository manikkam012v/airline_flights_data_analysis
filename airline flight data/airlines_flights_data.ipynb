```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load the Data File
data=pd.read_csv('airlines_flights_data.csv')
```

1.showing all the airline with their number of flights in the bar

```python
#Remove the 'index' column
data.drop(columns='index',inplace=True)
```

```python
data
```

```python
#Get some info about the data set
data.info()
```

```python
#data statistical summery about the dataset
data.describe()
```

```python
#checking how many airlines are in the dataset
data['airline'].nunique()
```

```python
# showing the name of the airline in the dataset
data['airline'].unique()
```

```python
#showing the all airline with their frequeince
data['airline'].value_counts()
```

```python
# showing all the airline with their number of flights in the bar
data['airline'].value_counts(ascending=True).plot.bar(color=['#9ECAD6','#F5CBCB','#748DAE'])
plt.title('airlines with frequencies')
plt.xlabel('airlines')
plt.ylabel('number of flights')
plt.show
```

2.show Bar graphs representing the Depature and Arrivale time

```python
data.head()
```

```python
#showing the departure time for the flights
data['departure_time'].value_counts()
```

```python
#showing the arrival time for the flights
data['arrival_time'].value_counts()
```

```python
#show the depature time and arrival time for the flight with their counts
plt.Figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.bar(data['departure_time'].value_counts().index,data['departure_time']
        .value_counts().values,color=['#D2C1B6','#1B3C53'])
plt.title('departure time')
plt.xlabel('departure name')
plt.ylabel('departure frequeince ')
plt.subplot(1,2,2)
plt.bar(data['arrival_time'].value_counts().index,data['arrival_time']
        .value_counts().values,color=['#D2C1B6','#1B3C53'])
plt.title('arrival time')
plt.xlabel('arrival name')
plt.ylabel('arrival frequeince ')
plt.show()
```

3.show the bar graph representing the Source city & Destination city

```python
data.head()
```

```python
#showing the source city of the flights
data['source_city'].value_counts()
```

```python
#showing the destination city of the flights
data['destination_city'].value_counts()
```

```python
#showing the source city & destination city for the flights with their counts
plt.Figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.bar(data['source_city'].value_counts().index,data['source_city'].value_counts().values,
        color=['#4DFFBE','#63C8FF'])
plt.title('source cities with no of flights')
plt.xlabel("cities")
plt.ylabel('no of flights')
plt.subplot(1,2,2)
plt.bar(data['destination_city'].value_counts().index,data['destination_city'].value_counts().values,
        color=['#4DFFBE','#63C8FF'])
plt.title('destination cities with no of flights')
plt.xlabel("cities")
plt.ylabel('no of flights')
plt.show()
```

4.Does Price varies with airlines?

```python
data.head()
```

```python
#Grouping the airlines and chencking the mean price
data.groupby('airline')['price'].mean()
```
